# Brüche 

$$
\Large{

\frac{Numerator}{Denominator}

}
$$

---
### 1.Addition

Um zwei Brüche zu addieren müssen wir sie zunächst gleichnamig machen, das bedeutet sie auf den selben Zähler (*Denominator*) zu bringen, der einfachste Weg ist die Brüche mit dem Zähler des jeweils anderen zu erweitern. 

$$

\Large{

\frac{a}{b} + \frac{c}{d}  = \frac{ad}{bd} + \frac{cb}{db} = \frac{ad+cb}{bd}
}
$$

Nehmen wir zum Bespiel $\Large{\frac{2}{3}}$ und $\Large{\frac{5}{2}}$:

$$
\Large{

\frac{2}{3}+\frac{5}{2}

}
$$

gleichnamig machen: 
$$
\Large{

\frac{4}{6}+\frac{15}{6}

}
$$

Und die *Numeratoren* addieren: 

$$
\Large{

\frac{4+15}{6} = \frac{19}{6}

}
$$

Vereinfachen: 

$$
\Large{

\frac{19}{6} = 3\frac{1}{6}

}
$$



---

### 2.Subtraktion

Auch hier müssen wir einen gemeinsamen Nenner finden, bevor wir einfach die *Numeratoren* von einander abziehen.


$$

\Large{

\frac{a}{b} - \frac{c}{d}  = \frac{ad}{bd} - \frac{cb}{db} = \frac{ad-cb}{bd}
}
$$


Weil ich unglaublich Faul bin benutzen wir wieder $\Large{\frac{2}{3}}$ und $\Large{\frac{5}{2}}$ als Beispiel damit ich copy-pasten kann.

$$
\Large{

\frac{2}{3}-\frac{5}{2}

}
$$


gleichnamig machen: 
$$
\Large{

\frac{4}{6}-\frac{15}{6}

}
$$

subtraction: 

$$
\Large{

\frac{4-15}{6} = -\frac{11}{6}

}
$$

Vereinfachen: 

$$
\Large{

\frac{11}{6} = -1\frac{5}{6}

}
$$

---

### 3.Multiplication 

Multiplication ist einfach wir multiplizieren einfach $numerator$ mit $numerator$ und $denominator$ mit $denominator$. 

$$
\Large{

\frac{a}{b} \cdot  \frac{c}{d} =\frac{a\cdot c}{b\cdot d} 

}

$$

Und wir benutzen wieder $\Large{\frac{2}{3}}$ und $\Large{\frac{5}{2}}$:


$$
\Large{

\frac{2}{3} \cdot \frac{5}{2}

}
$$

$$
\Large{

\frac{2\cdot 5}{3\cdot 2}  = \frac{10}{6} = 1\frac{4}{6}

}

$$

--- 
### 4.Division 

Ganz einfach mit dem Kehrwert multiplizieren. 

$$
\Large{

\frac{a}{b} :  \frac{c}{d} = \frac{a}{b}  \cdot \frac{d}{c}

}

$$

Und nochmal $\Large{\frac{2}{3}}$ und $\Large{\frac{5}{2}}$:


$$
\Large{

\frac{2}{3} : \frac{5}{2} = \frac{2}{3} \cdot \frac{2}{5}

}
$$

$$
\Large{

\frac{2\cdot 2}{3\cdot 5}  = \frac{4}{15}

}

$$

```python
import math
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Markdown as md
import numpy as np
```

#### Konstantenregel: 
$$
\Large{
\begin{align*} 
&f(x)=1\\
&f'(x)=0 \\
\end{align*}
}
$$


#### Faktorregel 

$$
\Large{
\begin{align*} 
&y = C \cdot f(x)\\
&y' = C \cdot f'(x) \\
\end{align*}
}
$$

#### Potenzregel 

$$
\Large{
\begin{align*} 
&f(x) = x^n \\
&f'(x) = n \cdot x^{n-1}\\
\end{align*}
}
$$

#### Summenregel 

$$
\Large{
\begin{align*} 
&f(x) = f_1(x) + f_2(x)\\
&f'(x) = f'_1(x). + f'_2(x)\\
\end{align*}
}
$$

#### Produkte 

$$
\Large{
\begin{align*} 
&y = u \cdot v \\
&y' = u' \cdot v + v' \cdot u
\end{align*}
}
$$


#### Quotientenregel 

$$
\Large{
\begin{align*} 
&f(x) = u:v \\
&f'(x) = \frac{u'\cdot v - u \cdot v'}{v^2}
\end{align*}
}
$$

```python
def get_function_value(x) -> float:
    return math.pow(x,3)

def get_deriviatve(x) -> float:
    return 3*math.pow(x,2)

x_values = [x for x in range(-99,100,1)]
y_values = [get_function_value(x) for x in x_values]
y_hat_value = [get_deriviatve(x) for x in x_values]


plt.plot(x_values, y_values)
plt.plot(x_values, y_hat_value)

plt.grid(linestyle='-', linewidth=0.5)

plt.show()
```

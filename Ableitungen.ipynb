Bei einer Ableitung handelt es sich um eine Funktion, die die Steigung einer Funktion angibt.

Ist die Funktion $f(x)$ geben so nennen wir die Ableitung $f'(x)$ oder $\frac{d}{dx}f(x)$ (Leibnitz Notation). 


```python
def get_function_value(x) -> float:
    return math.pow(x,3)

def get_deriviatve(x) -> float:
    return 3*math.pow(x,2)

x_values = [x for x in range(-99,100,1)]
y_values = [get_function_value(x) for x in x_values]
y_hat_value = [get_deriviatve(x) for x in x_values]


plt.plot(x_values, y_values)
plt.plot(x_values, y_hat_value)

plt.grid(linestyle='-', linewidth=0.5)

plt.show()
```

## Ableitungsregeln


### Konstantenregel
$$
\Large{
\begin{align*} 
&f(x)=1\\
&f'(x)=0 \\
\end{align*}
}
$$


### Potenzregel 
Für eine Funktion $f$  mit $f(x) == x^r, r \in\mathbb{R}|\{0\}$ gilt:
$$
\Large{
\begin{aligned} 
f'(x) = r\cdot x^{r-1}
\end{aligned}
}
$$


### Faktorregel

Für eine Funktion $f$ mit $f(x) = c \cdot g(x), c\in \mathbb{R}$ gilt:

$$
\Large{
\begin{aligned} 
f'(x) = c\cdot g'(x)
\end{aligned}
}
$$



### Summenregel 
Für eine Funktion $f$ mit $f(x) = g(x) + h(x)$ gilt:
$$
\Large{
\begin{aligned} 
f'(x) = g'(x) + h'(x)
\end{aligned}
}
$$


### Kettenregel 

$$
\Large{
\begin{aligned} 
f(x) = u(v(x)) \newline
f'(x) = u'(v(x))
\end{aligned}
}
$$

### Quotientenregel 

$$
\Large{
\begin{align*} 
&f(x) = \frac{u}{v} \\
&f'(x) = \frac{u'\cdot v - u \cdot v'}{v^2}
\end{align*}
}
$$

### $\sin{x} \textrm{ und } \cos{x}$

$$
\Large{
\begin{aligned} 
\sin{x} \rightarrow \cos{x} \rightarrow -\sin{x} \rightarrow -\cos{x} \rightarrow \sin{x}
\end{aligned}
}
$$

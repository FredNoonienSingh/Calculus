# Markdown Guide

## Überschriften 

Für Überschriften setzen wir einfach “# ” vor die Zeile, umso mehr “#” desto kleiner die Überschrift. 


### smaller Heading 
#### even smaller Heading 
##### super small Heading 
###### smallest Heading 

---


## Hervorhebungen 

__Bold__ = ```__bold__```

*Italic* = ```*Italic*```

<mark>marked</mark> = ```<mark>marked</mark>```



---


## Zitate 

### Einfache Zitate:

> Schlaue und wichtige Sache
> <cite>- Wichtiger Mensch</cite>

### Nested:

> als “Wichtiger Mensch”, gesagt hat:
>> Schlaue und wichtige Sache

---


## Listen:


### Geordnet:

1. Step 1
2. Step 2
3. Step 3

### Ungeordnet:

- Step 
- Step
- Step 


#### Nesting:

1. Step 1
	1. Substep 
	2. substep
2. Step 2

- Step 
- Step
	- Step 
- Step 
- Step

__Ungeordnete Listen mit Zahlen__

- 12\. Wochen 
- 4\. Tage

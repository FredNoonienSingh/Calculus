# Latex Guide

## Basics 

Um $\LaTeX{}$ in euere Dokumente einzubinden braucht ihr nicht mehr zu tun, als den Text den ihr in $\LaTeX{}$ gerendert haben wollt in ```$``` einzuklammern. 
Einzelne Dollarzeichen ```$``` werden ‘inline’ gerendert. 
```$inline$``` = $inline$

Mit doppelten Dollarzeichen ```$$``` erzeugt ihr einen $\LaTeX{}$ Block: 
```$$block$$``` = $$Block$$

### Funktionen in $\LaTeX{}$

Um Formatierung zu steuern und bestimmte Symbole zu erzeugen, werden in $\LaTeX{}$ sogenannte Funktionen verwendend, diese haben das Format ‘\‘ (option+Shift+7 *(OSX)* /alt+shift+7 *(Win)* ), gefolgt von dem Namen der Function z.b. *sum*, *frac*, *cdot*, …
Darauf folgen geschweifte Klammern ‘{}’ für die Argumente, die Menge der Klammern, kann je nach Funktion variieren, ‘*\cdot*’ benötigt keine, ‘*\frac*’ hingegen 2. 

--- 


### Häufig verwendete Funktionen 

### ```\frac{}{}```

Erzeugt einen Bruch:

$$
\frac{denominator}{numerator}
$$


### ```\sum```

Erzeugt eine Summe:

$$
\sum{n}
$$

Um Schrittgröße und Grenzwert zu definieren können ```_```und ```^``` verwendet werden, 
als Bespiel würde "\sum_{n=1}^{\infty}{\frac{1}{n^s}}" zu   

$$
\sum_{n=1}^{\infty}{\frac{1}{n^s}}
$$

evaluiert werden.

## Formatierung 

### Größe

###### \tiny{} erzeugt: 
$$
\tiny{
x^2+y^2=r
}
$$

###### \scriptsize{} erzeugt: 

$$
\scriptsize{
x^2+y^2=r
}
$$


###### \small{} erzeugt: 

$$
\small{
x^2+y^2=r
}
$$

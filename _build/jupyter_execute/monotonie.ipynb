### __Monotoniesatz__:

Gegeben ist eine im Intervall $I$ differenzierbare Function $f$. 
Wenn für alle $x$ aus $I$:

Wachsend
- $f'(x) > 0$ gilt, dann ist $f$ streng monoton wachsend in $I$.
- $f'(x) \geq 0$ gilt, dann ist $f$ monoton wachsend in $I$.

Fallend
- $f'(x) < 0$ gilt, dann ist $f$ streng monoton fallend in $I$. 
- $f'(x) \leq 0$ gilt, dann ist $f$ monoton fallend in $I$. 


### __Satz__:

__Ist $f$ in $I$ zweimal differnzierbar, so kann man den Monotoniesatz auf $f'$ anwenden.__

Es gilt:

Wenn $f'(x) > 0$ für alle $x \in I$ gilt, dann ist der Graph von $f$ *linksgekrümmt*. 

Wenn $f'(x) < 0$ für alle $x \in I$ gilt, dann ist der Graph von $f$ *rechtsgekrümmt*.

#### Bespiel: 


$$
\large{
\begin{align*} 
&f(x) = x^2 \\
&f'(x) = 2x \\
&f''(x) = 2 \\
\end{align*}
}
$$

Gesucht ist das Monotonie verhalten in den Intervallen $[0;10]$ und $[-5; -1]$



```python
def get_function_value(x) -> float:
    return math.pow(x,2)

def get_derivative(x) -> float:
    return 2*x

def get_second_derivative(x):
    return(2)

cartesean_plot(funcs=[
    {'func':get_function_value, 'title': "$f(x)$"},
    {'func':get_derivative, 'title': "$f'(x)$"},
    {'func':get_second_derivative, 'title': "$f''(x)$"}
    ],
    lim=100)
```

Für $x \in (0;10]$ ist der Graph $f(x)$ streng monoton wachsend weil alle $f'(x)\textrm{-Werte}$ oberhalb der $y$-Achse sind und damit $f'(x) > 0$ gilt.

Alle $f(x)\textrm{werte}$ im Intervall $x \in[0;10]$ sind $\leq 0$ also (nur) monoton wachsend.

Da $f''(x)$ immer $2$ ist, ist $f''(x)$ immer größer als $0$ damit gilt: 

$f''(x) > 0 \rightarrow  \textrm{Linkskurve}$

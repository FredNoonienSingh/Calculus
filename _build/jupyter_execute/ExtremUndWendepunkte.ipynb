#### Maximum und Minimum 
##### *Satz*:
Eine innere Stelle $x_0 \in I$ ist eine __Maximumstelle__ von $f$, wenn

$\rightarrow f'(x_0)= 0$ ist und $f’$ an der Stelle $x_0$ einen Vorzeichenwechsel von $+$ nach $-$ hat oder 

$\rightarrow f'(x_0) = 0$ und $f''(x_0)< 0$ ist. 

Eine innere Stelle ist $x_0 \in I$ ist eine Minimumstelle von $f$, wenn

$\rightarrow f'(x_0) = 0$ und $f’$ an der Stelle $x_0$ einem Vorzeichenwechsel von $-$ nach $+$ hat oder

$\rightarrow f'(x_0) =0$ und $f''(x_0) > 0$ ist.

#### Wendestelle 
##### *Satz*:
Eine innere Stelle $x_0 \in I$ ist eine __Wendestelle__ von $f$, wenn:

$\rightarrow f''(x_0) =0$ ist und $f’’$ an der Stelle $x_0$, einen Vorzeichenwechsel hat oder 
 
$\rightarrow f''(x_0) =0$ und $f’’'(x_0) \neq 0$ ist.

#### a.

$$ f(x) = x^4-4x^3 $$

Um die Extrempunkte zu finden benötigen wir zunächst die erste Ableitung $f'(x)$

$$\frac{d}{dx}f(x) = 4x^3 - 12x^2 $$

Und für die Wendepunkte $f''(x)$:

$$\frac{d^2}{d^2 x}f(x) = 12x^2 - 24x$$


```python
def get_function_value(x) -> float:
    return math.pow(x,4) - 4*math.pow(x,3)

def get_derivative(x) -> float:
    return 4*math.pow(x,3) - 12*math.pow(x,2)

def get_second_derivative(x) -> float:
    return 12*math.pow(x,2) - 24*x

def get_third_derivative(x) -> float:
    return 24*x - 24

cartesean_plot(funcs=[
    {'func':get_function_value, 'title': "$f(x)$"},
    {'func':get_derivative, 'title': "$f'(x)$"},
    #{'func':get_second_derivative, 'title': "$ f''(x)$"},
    #{'func':get_third_derivative, 'title': "$ f'''(x)$"}
    ],
    lim=30)
```

### Extrempunkte 

Wir wissen, das ein Extrempunkt auf $\frac{d}{dx}f(x) = 0$ also:

$$
4x^3 - 12x^2 = 0 \\
4x^2(x - 3) = 0 \\
$$

Die $Nullproduktregel$ sagt uns, das wenn einer der Faktoren $0$ ist, das Produkt $0$ sein muss. Damit ergeben sich die Lösungen: 

$x = 0$ weil $4\cdot 0^2 = 0$ und $x = 3$ denn $(3-3) = 0$


Daraus ergeben sich die Extrempunkte $P(0|f(0))$ und $Q(3|f(3))$. 
Um diese Punkte nun als Hoch oder Tiefpunkt zu identifizieren müssen wir nun nach Vorzeichenwecheln in der ersten Ableitung suchen, also setzen wir für Punkt $P$ in die Ableitung $-1$ und $1$ und überprüfen ob sich das Vorzeichen verändert.

Am Punkt $P$ findet kein Vorzeichenwechel auf der ersten Ableitung statt, daher ist er kein Extrempunkt, ob es sich dabei um einen Wendepunkt handelt überprüfen wir zu einem Späteren Zeitpunkt. 
Bei Punkt $Q$ allerdings wechlelt das Vorzeichen von $-$ zu $+$, daher liegt hier ein $\textrm{Tiefpunkt}$ vor. 



### Wendestellen und Sattelpunkte

Bevor wir uns mit der zweiten Ableitung auf die Such nach Wendestellen machen, überprüfen wir zu erst $P$:

Für die Überprüfung benötigen wir $$\frac{d^3}{d^3 x}f(x) = 24x-24$$

Ein Wendepunkt liegt for wenn:

$\rightarrow \frac{d^2}{d^2 x}f(x_0) =0$ ist und $\frac{d^2}{d^2 x}$ an der Stelle $x_0$, einen Vorzeichenwechsel hat oder 
 
$\rightarrow \frac{d^2}{d^2 x}f(x_0) =0$ und $\frac{d^3}{d^3 x}f(x_0) \neq 0$ ist.

#### Untersuchung auf weitere Wendepunkte: 

$$
12x^2-24x = 0 \\

$$

Hier kommt wieder die $\textrm{Nullproduktregel}$ zum Einsatz, und damit erhalten wir: 

$x=0$ und $x=2$

Den ersten Fall haben wir bereits abgehandelt, also bleibt zur Überprüfung nur $x=2$
